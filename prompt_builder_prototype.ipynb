```python
import json
import random

json_str = '''{
  "Tone": ["playful", "whimsical", "lighthearted", "joyful", "mischievous"],
  "Tempo/Intensity": ["moderate", "lively", "energetic", "gentle", "varying"],
  "Instrumentation": ["flute", "xylophone", "ukulele", "clarinet", "pizzicato strings"],
  "Related Music Setting/Music Genre": ["Children's Music", "Light Acoustic", "Fantasy Soundtrack", "Comedy", "Cartoon"],
  "Transition/Foreshadow": ["anticipatory", "building", "sudden shifts", "playful pauses", "dynamic changes"],
  "Description/Adjectives of Scene": ["colorful", "dynamic", "animated", "intriguing", "engaging"],
  "Pitch": ["high", "bright", "varied", "melodic", "harmonic"],
  "Crescendo": true
}
'''

json_str_2 = '''{
"tone": "calm",
"intensity": "medium",
"instrumentation": "piano",
"setting": "outdoors",
"transition": "smooth",
"scene_description": "a peaceful forest",
"pitch": "middle",
"crescendo": false
}
'''

json_str_3 = '''{
"tone": "Dark",
"intensity": "Loud",
"instrumentation": "Ethereal",
"setting": "Dark, ominous",
"scene_description": "A dark figure races towards the bridge",
"pitch": "C",
"crescendo": true
}
'''

# def get_element(key, value):
#     if isinstance(value, list):
#         return str(random.choice(value))
#     if isinstance(value, bool) and key.lower() == 'crescendo':
#         return "With crescendo" if value else "Without crescendo"
#     return str(value)

def get_element(value):
    if isinstance(value, list):
        return str(random.choice(value))
    if isinstance(value, bool):
        return value
    return str(value)
    


def info_to_prompt(json_str, SEP=', '):
    def get_element(value):
        if isinstance(value, list):
            return str(random.choice(value))
        if isinstance(value, bool):
            return value
        return str(value)
    
    key_order = {'scene_description': '{}',
             'setting': '{} setting',
             'related music setting/music genre': '{} music setting',
             'description/adjectives of scene': '{} description',
             'tone': '{} tone',
             'intensity': '{} intensity',
             'tempo/intensity': '{} tempo/intensity',
             'instrumentation': '{} instrumentals',
            #  'transition': '{} transition',
             'pitch': '{} pitch',
             'crescendo': '{}'
             }

    bool_to_text = {'crescendo': {True: 'with crescendo', False: ''}
                }
    
    info = json.loads(json_str)
    info = {k.lower(): v for k, v in info.items()}
    prompt = ""
    
    for key, format_str in key_order.items():
        if key in info:
            val_str = get_element(info[key]) # (key, info[key])
            if isinstance(val_str, bool) and key in bool_to_text:
                val_str = bool_to_text[key][val_str]
            if val_str != "":
                if prompt != "":
                    prompt += SEP
                prompt += f"{format_str.format(val_str)}"

    return prompt

print(info_to_prompt(json_str))
print(info_to_prompt(json_str))
print(info_to_prompt(json_str))

print(info_to_prompt(json_str_2))
print(info_to_prompt(json_str_3))



```
